```python
import pandas as pd
import numpy as np
import re
```

```python
f = open("Classes.nt", "w", encoding="utf-8")
```

# Retrieve the Data from LOV

```python
lov_df = pd.read_csv("./../0. Ontologies Data/LOV/all_classes_from_lov.csv", encoding="utf-8", names=["Class","Context","Label","Description"], skiprows=1).replace(np.nan, "")
lov_df
```

```python
def write_line_LOV(line, f):
    if line["Label"] != "":
        label = re.sub("\s", " ", line["Label"])
        label = label.replace('"'," ")
        label = label.replace('\\'," ")
        f.write(f'<{line["Class"]}> <http://www.w3.org/2000/01/rdf-schema#label> "{label}".\n')
    f.write(f'<{line["Class"]}> <http://graph/origin> <{line["Context"]}>.\n')
    f.write(f'<{line["Class"]}> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/1999/02/22-rdf-syntax-ns#Class>.\n')
    if line["Description"] != "":
        description = re.sub("\s", " ",line["Description"])
        description = description.replace('"'," ")
        description = description.replace('\\'," ")
        f.write(f'<{line["Class"]}> <http://www.w3.org/2000/01/rdf-schema#description> "{description}".\n')
```

```python
lov_df.apply(lambda x:write_line_LOV(x, f), axis=1)
""
```

# Retrieve the Data from Wikidata

```python
wk_df = pd.read_csv("./../0. Ontologies Data/Wikidata/all_classes_from_wikidata.txt", encoding="utf-8", names=["Class","Label","Description"], sep="\t").replace(np.nan, "")
wk_df
```

```python
def write_line_wikidata(line, f):
    if line["Label"] != "":
        label = re.sub("\s", " ", line["Label"])
        label = label.replace('"'," ")
        label = label.replace('\\'," ")
        f.write(f'<{line["Class"]}> <http://www.w3.org/2000/01/rdf-schema#label> "{label}".\n')
    f.write(f'<{line["Class"]}> <http://graph/origin> <http://wikidata.org>.\n')
    f.write(f'<{line["Class"]}> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/1999/02/22-rdf-syntax-ns#Class>.\n')
    if line["Description"] != "":
        description = re.sub("\s", " ", line["Description"])
        description = description.replace('"'," ")
        description = description.replace('\\'," ")
        f.write(f'<{line["Class"]}> <http://www.w3.org/2000/01/rdf-schema#description> "{description}".\n')
```

```python
wk_df.apply(lambda x:write_line_wikidata(x, f), axis=1)
""
```
