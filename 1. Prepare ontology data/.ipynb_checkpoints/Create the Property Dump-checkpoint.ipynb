```python
import pandas as pd
import numpy as np
import re
```

```python
f = open("Properties.nt", "w", encoding="utf-8")
```

# Retrieve the Data from LOV

```python
lov_df = pd.read_csv("./../0. Ontologies Data/LOV/all_props_from_lov.csv", encoding="utf-8", names=["Property","Type","Context","Range","Label","Description", "Domain"], skiprows=1).replace(np.nan, "")
lov_df["Domain"] = lov_df["Domain"].map(lambda x: x[2:-2].split("', '"))
lov_df["Range"] = lov_df["Range"].map(lambda x: x[2:-2].split("', '"))
lov_df
```

```python
def write_line_LOV(line, f):
    if line["Label"] != "":
        label = re.sub("\s", " ", line["Label"])
        label = label.replace('"'," ")
        label = label.replace('\\'," ")
        f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#label> "{label}".\n')
    f.write(f'<{line["Property"]}> <http://graph/origin> <{line["Context"]}>.\n')
    f.write(f'<{line["Property"]}> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <{line["Type"]}>.\n')
    f.write(f'<{line["Property"]}> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/1999/02/22-rdf-syntax-ns#Property>.\n')
    if line["Description"] != "":
        description = re.sub("\s", " ", line["Description"])
        description = description.replace('"'," ")
        description = description.replace('\\'," ")
        f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#description> "{description}".\n')
    if line["Domain"]!=['']:
        for d in line["Domain"]:
            if d[:4] == "http":
                f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#domain> <{str(d)}>.\n')
    if line["Range"]!=['']:
        for r in line["Range"]:
            if r[:4] == "http":
                f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#range> <{str(r)}>.\n')
```

```python
lov_df.apply(lambda x:write_line_LOV(x, f), axis=1)
""
```

# Retrieve the Data from Wikidata

```python
wk_df = pd.read_csv("./../0. Ontologies Data/Wikidata/all_props_from_wikidata.txt", encoding="utf-8", names=["Property","Label","Description","Wikibase domain","Range", "Domain"], sep="\t").replace(np.nan, "")
wk_df
```

```python
def write_line_wikidata(line, f):
    if line["Label"] != "":
        label = re.sub("\s", " ", line["Label"])
        label = label.replace('"'," ")
        label = label.replace('\\'," ")
        f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#label> "{label}".\n')
    f.write(f'<{line["Property"]}> <http://graph/origin> <http://wikidata.org>.\n')
    f.write(f'<{line["Property"]}> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2000/01/rdf-schema#Property>.\n')
    if line["Description"] != "":
        description = re.sub("\s", " ", line["Description"])
        description = description.replace('"'," ")
        description = description.replace('\\'," ")
        f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#description> "{description}".\n')
    if line["Domain"] != " ":
        for d in line["Domain"].lower().split(","):
            if d[:4] == "http":
                f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#domain> <{d}>.\n')
    if line["Range"] != " ":
        for r in line["Range"].lower().split(","):
            if r[:4] == "http":
                f.write(f'<{line["Property"]}> <http://www.w3.org/2000/01/rdf-schema#range> <{r}>.\n')
```

```python
wk_df.apply(lambda x:write_line_wikidata(x, f), axis=1)
""
```

```python
f.close()
```
